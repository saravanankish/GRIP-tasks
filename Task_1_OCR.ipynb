Importing necessary libraries

```
from PIL import Image
import pytesseract
import os
from google.colab.patches import cv2_imshow
import cv2
```

Function to extract text from the Image using pytesseracct 

```
def OCR(image_path):
  img = Image.open(image_path)
  image = cv2.imread(image_path)
  cv2_imshow(image)
  string = pytesseract.image_to_string(img, lang='eng')
  print(string)
```

Text Extracted from handwritten text image1

```
OCR('test1.jpg')
```
