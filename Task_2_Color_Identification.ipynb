# **Extraction of given number of color in hex format**

Importing Necessary modules

```
import cv2
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from collections import Counter
import numpy as np
```

Function to convert BGR format to RBG format

```
def BGR2RGB(image):
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image
```

Function to convert RGB format to HEX format

```
def RGB2HEX(color):
    return "#{:02x}{:02x}{:02x}".format(int(color[0]), int(color[1]), int(color[2]))
```

Function to get the image from the path

```
def get_image(path):
    image = cv2.imread(path)
    image = BGR2RGB(image)
    return image
```

Function to extract given number of colors from the image path. It uses KMeans algorithm to find the given number of colors in HEX format

```
def detect_colors(path, number_of_colors):
    image = get_image(path)
    plt.imshow(image)
    image = cv2.resize(image, (600, 600), interpolation=cv2.INTER_AREA)
    image = image.reshape(image.shape[0] * image.shape[1], image.shape[2])
    k_means = KMeans(n_clusters=number_of_colors)
    labels = k_means.fit_predict(image)
    counts = Counter(labels)
    center_colors = k_means.cluster_centers_
    ordered = [center_colors[i] for i in counts.keys()]
    hex_colors = [RGB2HEX(ordered[i]) for i in counts.keys()]

    plt.figure(figsize=(8, 6))
    plt.pie(counts.values(), labels=hex_colors, colors=hex_colors)
    plt.show()
```

Extraction of given set of colors from image 1

```
detect_colors(input("Enter the path: "), int(input("Enter No.of colors to be detected: ")))
```

Extraction of given set of colors from image 2

```
detect_colors(input("Enter the path: "), int(input("Enter No.of colors to be detected: ")))
```
